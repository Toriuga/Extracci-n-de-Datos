```python
#importamos librerioas requeridas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Verificamos información del DataFrame
df.info()
```

```python
#Filtro por objeto
filtro1=df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro1

#Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro por objeto.csv")

#filtra en la columna nombre_vendedor, el nombre de leticia ramirez
```

```python
#Filtro por fila 
filtro2= df.iloc[3:10,: ] 
filtro2

#Convertir archivo filtrado a CSV
filtro2.to_csv("Filtro por fila .csv")

#hace un filtro que solo muestra de la fila 3 a la fila 9 (la primera son los indices)
```

```python
#Filtro por columna 
filtro3=df.iloc[ : , 1:5]  
filtro3

#Convertir archivo filtrado a CSV
filtro3.to_csv("Filtro3 por columna .csv")

#filtra de la columna 1 a la 5
```

```python
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1

#Aplicamos filtro de filas vs columnas
filtro4=df1.loc[["ALFREDO CANELA","LETICIA RAMIREZ HERNANDEZ"], ["NOMBRE_CLIENTE"]]
filtro4

#Convertir archivo filtrado a CSV
filtro4.to_csv("filtro de filas vs columnas.csv")

#muestra solamente en donde el campo nombre vendedor sea alfredo canela o leticia ramirez y muestra el campo del vendedor y campo del cliente
```

```python
#filtro por cabecera
filtro5=df.head()
filtro5

#Convertir archivo filtrado a CSV
filtro5.to_csv("filtro por cabecera.csv")

#Filtra y muestra las primeras 5 filas
```

```python
#Filtro por cola
filtro6=df.tail()
filtro6

#Convertir archivo filtrado a CSV
filtro6.to_csv("Filtro por cola.csv")

#muestra las ultimas 5 filas
```

```python
#Filtro por comparacion
filtro7=df[df["COSTO_TOTAL_CALCULADO"] > 1000]
filtro7

#Convertir archivo filtrado a CSV
filtro7.to_csv("Filtro por comparacion.csv")

#filtra las filas donde la columna COSTO_TOTAL_CALCULADO sea mayor a 1000
```

```python
#Filtro Y
filtro8=df[(df["COSTO_TOTAL_CALCULADO"] < 1000) & (df["FECHA_DOC"] == "2022-01-05")]
filtro8

#Convertir archivo filtrado a CSV
filtro8.to_csv("Filtro Y.csv")

#filtra los valores columna COSTO_TOTAL_CALCULADO donde sea menor a 1000 y la columna FECHA_DOC sea igual a "2022-01-05"
```

```python
#Filtro o
filtro9=df[(df["COSTO_TOTAL_CALCULADO"] < 1000) | (df["FECHA_DOC"] == "2022-10-01")]
filtro9

#Convertir archivo filtrado a CSV
filtro9.to_csv("Filtro o.csv")

#filtra los valores columna COSTO_TOTAL_CALCULADO donde sea menor a 1000 o la columna FECHA_DOC sea igual a "2022-10-01"
```

```python
#Filtro not
filtro10=df[~(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro10

#Convertir archivo filtrado a CSV
filtro10.to_csv("Filtro not.csv")

#filtran los valores donde la columna NOMBRE_VENDEDOR no sea igual a LETICIA RAMIREZ HERNANDEZ
```
